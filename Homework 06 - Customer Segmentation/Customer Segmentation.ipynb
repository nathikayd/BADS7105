```
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline 
from sklearn.cluster import KMeans
import numpy as np

mpl.font_manager.fontManager.addfont('/content/drive/MyDrive/CRM/K2D-Regular.ttf')
mpl.rc('font', family='K2D', size=12, weight=200)
```

```
filename = 'Supermarket Data.csv'
df = pd.read_csv(filename)
df.head(10)
```

```
df = df.groupby('CUST_CODE').agg(
        total_visit=('BASKET_ID', pd.Series.nunique),                                        
        total_spend=('SPEND', sum), 
        )
    
```

```
df
```

**Run elbow method find the best K.**

```


wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init="k-means++")
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))    
plt.grid()
plt.plot(range(1,11),wcss, linewidth=2, color="red", marker ="8")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("WCSS")
plt.show()
```
